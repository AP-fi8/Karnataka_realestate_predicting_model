```python
## here we build a function which can remove commas from payment_mar column values
```

```python
preprocessed_df['payment_mar']=preprocessed_df['payment_mar'].astype(str)

def remove_comma(val):
    token=val.split(',')
    if len(token)==3:
        a=token[0]
        b=token[1]
        c=token[2]
        return int(a+b+c)
    else:
        return int(val)
preprocessed_df['payment_mar']=preprocessed_df['payment_mar'].apply(remove_comma)
```

```python
## here we built a function which removes apostrophe from the payment_apr column values
## then we use the previous remove_comma function to removes commas
```

```python
preprocessed_df['payment_apr']=preprocessed_df['payment_apr'].astype(str)

def remove_apostrophe(val):
    a=val.replace("'", "")
    return a

preprocessed_df['payment_apr']=preprocessed_df['payment_apr'].apply(remove_apostrophe)
```

```python
def remove_comma(val):
    token=val.split(',')
    if len(token)==2:
        a=token[0]
        b=token[1]
        return int(a+b)
    else:
        return int(val)

preprocessed_df['payment_apr']=preprocessed_df['payment_apr'].apply(remove_comma)
```

```python
## d=due date of payment
## d+3= payments done three days after due date
## d+6= payments done 6 days after due date
## d+15= payments done 15 days after due date
## d-3=payments done 3 days before due date
```

```python
## payments on time if payments done atmost 5 days after due date
## otherwise late
```

```python
## so d+6 and d+15 are late ones(0)
## rest are on time(1)
```

```python
preprocessed_df['bucket_jan']=preprocessed_df['bucket_jan'].map({'D':1,'D+3':1,'D-3':1,'D+6':0,'D+15':0})
preprocessed_df['bucket_feb']=preprocessed_df['bucket_feb'].map({'D':1,'D+3':1,'D-3':1,'D+6':0,'D+15':0})
preprocessed_df['bucket_mar']=preprocessed_df['bucket_mar'].map({'D':1,'D+3':1,'D-3':1,'D+6':0,'D+15':0})
preprocessed_df['bucket_apr']=preprocessed_df['bucket_apr'].map({'D':1,'D+3':1,'D-3':1,'D+6':0,'D+15':0})
preprocessed_df['bucket_may']=preprocessed_df['bucket_may'].map({'D':1,'D+3':1,'D-3':1,'D+6':0,'D+15':0})
```

```python
division_dummy=pd.get_dummies(preprocessed_df['Division'],drop_first=True)
## creating dummies
division_dummy.head()
```

```python
preprocessed_df=pd.concat([preprocessed_df,division_dummy],axis=1)
```

```python
## based on this we divide the column in to two binary classes, where
## 0 denotes late payments, and otherwise one
## here 0,1,2 are late ones, and rest are on time
```

```python
payment_pattern=payment_pattern.map({0:0,1:0,2:0,3:1,4:1,5:1})
preprocessed_df['payment_behavior']=payment_pattern
```

```python
## selecting inputs for logistic reg
unscaled_inputs=preprocessed_df.iloc[:,1:-1]
```

```python
## standarization
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin):
    def  __init__(self,columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self,X,y = None):
        self.scaler.fit(X[self.columns],y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self,X,y = None,copy = None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]),columns = self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled],axis = 1)[init_col_order]
```

```python
unscaled_inputs.columns

columns_to_scale=['payment_jan', 'payment_feb', 'payment_mar', 'payment_apr',
       'payment_may']
df_scaler=CustomScaler(columns_to_scale)
df_scaler.fit(unscaled_inputs)
scaled_inputs=df_scaler.transform(unscaled_inputs)
scaled_inputs.head()
```

```python
## to gain interpretation of categorical vatriables we drop this column
scaled_inputs.drop('bucket_jan',axis=1,inplace=True)
```
